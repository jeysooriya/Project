```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Tesla Stock Data
tesla = yf.Ticker("TSLA")
tesla_data = tesla.history(period="max")
tesla_data.reset_index(inplace=True)
tesla_data.head()
```

```python
# GameStop Stock Data
gme = yf.Ticker("GME")
gme_data = gme.history(period="max")
gme_data.reset_index(inplace=True)
gme_data.head()
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Step 1: Target the Tesla revenue page from Macrotrends
url = "https://www.macrotrends.net/stocks/charts/TSLA/tesla/revenue"

# Step 2: Use headers to avoid 403 Forbidden error
headers = {
    "User-Agent": "Mozilla/5.0"
}

# Step 3: Make the request
response = requests.get(url, headers=headers)

# Step 4: Parse the HTML content
soup = BeautifulSoup(response.text, 'html.parser')

# Step 5: Find all tables on the page
tables = soup.find_all("table")

# Step 6: Filter the correct table with "Tesla Quarterly Revenue"
for table in tables:
    if "Tesla Quarterly Revenue" in table.text:
        tesla_revenue = pd.read_html(str(table))[0]
        break

tesla_revenue.columns = ["Date", "Revenue"]
tesla_revenue["Revenue"] = tesla_revenue["Revenue"].str.replace("$", "").str.replace(",", "")

tesla_revenue.tail()
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL for GME Revenue on Macrotrends
url = "https://www.macrotrends.net/stocks/charts/GME/gamestop/revenue"

# Fake browser header to bypass 403 Forbidden
headers = {"User-Agent": "Mozilla/5.0"}

# Send request
response = requests.get(url, headers=headers)

# Parse HTML
soup = BeautifulSoup(response.text, 'html.parser')
tables = soup.find_all('table')

# Find the table containing "GameStop Quarterly Revenue"
for table in tables:
    if "GameStop Quarterly Revenue" in table.text:
        gme_revenue = pd.read_html(str(table))[0]
        break
# Clean the dataframe
gme_revenue.columns = ["Date", "Revenue"]
gme_revenue["Revenue"] = gme_revenue["Revenue"].str.replace("$", "").str.replace(",", "")
gme_revenue = gme_revenue[gme_revenue["Revenue"] != ""]
gme_revenue.tail()
```

```python
import plotly.graph_objects as go

# Filter data from 2012 onwards
tesla_data_2012 = tesla_data[tesla_data['Date'].dt.year >= 2012]

fig = go.Figure()

# First trace: Tesla Close Price
fig.add_trace(go.Scatter(
    x=tesla_data_2012['Date'],
    y=tesla_data_2012['Close'],
    mode='lines',
    name='Tesla Close Price',
    line=dict(color='blue')
))

# Second trace: Tesla Volume (for contrast)
fig.add_trace(go.Scatter(
    x=tesla_data_2012['Date'],
    y=tesla_data_2012['Volume'],
    mode='lines',
    name='Volume',
    yaxis='y2',
    line=dict(color='gray')
))

# Update layout with dual y-axes and range slider
fig.update_layout(
    title='Tesla Stock Price and Volume (Since 2012)',
    xaxis=dict(title='Date', rangeslider=dict(visible=True)),
    yaxis=dict(title='Closing Price (USD)', side='left'),
    yaxis2=dict(title='Volume', overlaying='y', side='right', showgrid=False),
    legend=dict(x=0, y=1),
    height=500
)

fig.show()
```

```python
import plotly.graph_objects as go

# Filter data from 2006 onwards
gme_data_2006 = gme_data[gme_data['Date'].dt.year >= 2006]

fig = go.Figure()

# First trace: Close Price
fig.add_trace(go.Scatter(
    x=gme_data_2006['Date'],
    y=gme_data_2006['Close'],
    mode='lines',
    name='GME Close Price',
    line=dict(color='darkred')
))

# Second trace: Volume (for visual contrast)
fig.add_trace(go.Scatter(
    x=gme_data_2006['Date'],
    y=gme_data_2006['Volume'],
    mode='lines',
    name='Volume',
    yaxis='y2',
    line=dict(color='gray')
))

# Update layout for dual y-axes and range slider
fig.update_layout(
    title='GameStop Stock Price and Volume (Since 2006)',
    xaxis=dict(title='Date', rangeslider=dict(visible=True)),
    yaxis=dict(title='Closing Price (USD)', side='left'),
    yaxis2=dict(title='Volume', overlaying='y', side='right', showgrid=False),
    legend=dict(x=0, y=1),
    height=500
)

fig.show()
```
